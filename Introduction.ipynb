# Introduction to Data Analysis 
<a id="introduction-to-data-analysis"></a>

Data analysis is the process of examining, organizing, transforming, and interpreting data to extract meaningful insights and support decision-making. Python has emerged as one of the most popular programming languages for data analysis due to its simplicity, versatility, and a rich ecosystem of tools designed specifically for handling data.
Python is capable of working with small datasets for quick analyses as well as massive datasets.

In the example of the sleep data analysis, we will work with data collected in a s survey, asking participants about their sleeping routines. The data is anonymised and brought together in an Excel file. From the Excel file, the data can be read in by a program written in Python and can transform the data into graphs, making it much easier to analyse and interpret the data.

When using Python for data analysis, the process typically involves the following steps:
#### Data Collection
Gathering data is the first step, which might involve accessing databases, downloading files, or collecting information from web services.

#### Data Cleaning
Raw data often contains inconsistencies, missing values, or errors. Cleaning involves identifying and rectifying these issues to ensure the data is ready for analysis.

#### Data Exploration
This step involves examining the data to understand its structure, patterns, and distributions. Techniques like summarizing statistics, visualizing data, and identifying trends help uncover insights.

#### Data Transformation
Sometimes, data needs to be reshaped or processed to make it suitable for analysis. This may include filtering, aggregating, or merging datasets.

#### Analysis and Interpretation
Applying statistical or computational methods to derive insights or test hypotheses. This may involve identifying correlations, trends, or anomalies.

#### Visualization
Representing data graphically makes it easier to understand and communicate findings. Charts, graphs, and other visual tools help to present complex data in an accessible format.

#### Reporting
The final step is to compile results and share them in a meaningful way, which may include written reports, dashboards, or presentations.

### Data types and variables 
<a id="data-types-and-variables"></a>

#### Data Types
In programming, one of the most important concepts are data types. Depending on the data we are working with, there are different types of data that we are going to encounter, and different data types can be used for to do different things.
In Python, there are a number of different data types. The most important ones are:

##### str
str stands for string and is used to store text. If we want to use hello as a string, it is important that we type the string between "" -> "hello". Now we have a hello string. A string can include numbers, letters or other characters, as long as we type the string between " ". We can even leave it blank and not type anything in between " ".

##### int 
int stands for integer and is used to store whole number that can be positive, negative or zero, for example 144. An integer cannot be a fractional number. 

##### float
A float is a decimal or fractional number, for example 22.5. A flot can be a postive or negative number.

##### bool
bool stands for boolean. This is a very interesting data type, because it can only hold one of two different values - either true or false. We will learn more about boolean variables and  what this data type is used for in the module for Loops.

##### list 
Lists can be used to store multiple items in one single variable, where each item is allocated an index number, which can referenced to refer to the item.

#### Variables
If we want to store data values, we can use variables. Variables act like containers and can store data. For example, we can create a variable y and assign the value 5 to it, y = 5. y now has the value 5.

If we want to change the value assigned to a variable, we can overwrite it by assigning a new value to it. y = 6 will now assign 6 to y, and overwrite y = 5. 

A variable can be a string, a float, an integer, a list or another data type.

### Lists
<a id="lists"></a>
Lists are a data type in Python and can be used to store multiple items in a single variable. Lists are created by using square brackets [] and listing the items on the list, for example:

> myList = ["apple", "banana", "cherry"]

It is important that the items in the list are listed in between " ". We can separate the items in the list through a comma. There is no limit to the number of items we can have in the list, and we can even delete or add items to the list later.
A list is always ordered, which means that the order will not change. If we add items to our list, they will be placed at the end of the list. A list also allows for duplicates, which means that we can add the same item to the list twice.
We can also add more items to the list by appending them, which is exlained in the Code Example section.

All lists in Python are indexed, which means that each item has an index number. The first item has index[0],the second item has index[1].  This is particularly important when we want to retrieve a specific item from the list.

Now we can add items to the list, print the list and even print the items for specific indices. There is also a functionality to delete items from the list or the entire list, which is explained in the Code Example section.

### Loops 
<a id="loops"></a>

Loops are constructs that allow us to repeat a block of code multiple times without having to write the code repeatedly. Loops help to automate repetitive tasks and iterate over collections of data. For example, if we have a list, we can use a loop to go through the different items in the list to access what is stored in the list. In Python, there are different types of loops:

#### for loop
A for loop can be used to iterate over a sequence, such as a list, tuple or string For example, if we have a list of items, we can use a for loop to print the items in the list.

For example, the following is a list named animals, which contains three different animals. a For loop allows us to loop through the list and print out each item in the list without having to call a print command for each item individually.

> animals = ["dog", "cat", "fish"]
> 
> for animal in animals:
> 
>       print(animal)

#### while loop
A while loop executes a block of code repeatedly as long as a specified condition is true. This is useful if we want a program to keep running and not finish after we run it once. We can define a certain condition that will determine whether our program will keep running. It we want our program to keep running indefinitely until we close it, we can set the condition to true. That means, as the condition won't ever be false (unless we change it), that our program will keep running.

For example, we create a variable num to hold a number. We start with 1. While num (the number) is less than 5, the program will print the number and add 1 to it - until it reaches 5.

> num = 5
> 
> while num <= 5:
> 
>       print(num)
> 
>        num += 1

First, we create a variable for age and assign an age to it.
Then we check whether the age is greather than or equal to 18.
If the age is greater than or equal to 18, the program should print a message "You are eligible to vote."
If the age is less than 18, the program should print a message "You are not eligible to vote."

### Generating data 
<a id="generating-data"></a>

In the context of sleep data analysis, Python programs can handle data stored in formats such as Excel spreadsheets, where data is read directly into the program for analysis. However, we can also use Python modules to generate random data, which allows us to experiment, test, or visualize concepts without relying on existing datasets.

For instance, imagine we want to simulate a survey to determine peopleâ€™s favorite pets. Instead of manually collecting or sourcing this data, Pythonâ€™s tools can help us create it automatically. This approach is especially helpful for practice or when working with theoretical scenarios where real-world data isn't available.

#### Creating a List of Pets
To simulate the survey, we first define the list of potential pet typesâ€”e.g., dogs, cats, fish, birds, and others. Each pet type represents a possible answer to our survey. By assigning this list to a variable, we create an easy way to reference and reuse it throughout our code. For example, if we want to adjust the survey to include new pet types (like rabbits or reptiles), we only need to update the list, and the rest of the program will adapt automatically.

#### Generating Random Data Using Python's Random Module
Python includes a powerful module called random, which allows us to generate random numbers and make random selections. This module is especially useful for creating synthetic datasets like the one needed for our survey simulation.

Each item in our list of pets has an index number, which represents its position in the list (e.g., the first item has an index of 0, the second an index of 1, and so on). By generating random numbers corresponding to these index values, the program can simulate random survey responses.

For example:

If the list is ["dog", "cat", "fish", "bird"], and the program randomly selects the index 2, it will choose "fish" as the response.
We can also simulate the number of people who selected each pet as their favorite. For instance:

If we want to simulate a range of responses from 0 to 100, the program will generate a random number within this range for each pet type.
This number could represent how many people chose a particular pet as their favorite in a hypothetical survey.
#### Benefits of Generating Random Data
##### Flexibility: 
Random data generation allows you to quickly test ideas, algorithms, or visualizations. <br> 

##### Scalability:
You can easily generate larger datasets by increasing the range of numbers or the list size. <br> 

##### Reproducibility: 
By setting a "random seed," you can ensure that the same random numbers are generated each time the program is run, which is helpful for debugging and sharing results.<br>

#### Practical Applications
The random data generated can then be used to create graphs and charts. For example:

A bar chart could display the number of people who selected each pet as their favorite.
A pie chart could visualize the proportion of each pet type in the survey results.
By leveraging Pythonâ€™s random module, we can simulate realistic datasets, analyze them, and present the results in a meaningful way, all without the need for real-world data collection. This process is invaluable for learning data analysis, prototyping, or demonstrating concepts.

### Pie charts 
<a id="pie-charts"></a>

Graphs, also known as charts or plots, are visual representations of data. They are widely used to analyze and present data in a graphical format. There are different types of graphs available - in this module we will look at two types of graphs: Pie Chart and and Bar Graph. 
Pie Charts are used to display data as sclices of a circular pie. This is useful for showing the proportion of each cateogory in a dataset.

In the sleep study, the data was collected from people participating in the survery. The data was stored and then read in by the program and displayed in a graph. In this example, we will simulate data collected in a survey.
In this example, we simulate a survey asking people about their favourite pets. We store the different pets in a list and then get the program to generate random numbers of how many people voted for each pet - between 0 and 100, simulating that 100 people participated in the survery. Every time the program is run, it generates new data.
Alternatively, we can create a static list with the numbers for each pet if we prefer to keep working with the same data.

We then display the percentage of people voting for each pet.

```python
import random
pets = ['Dogs', 'Cats', 'Birds', 'Fish', 'Rabbits']
#data = [18, 25, 72, 79, 68]

data = [random.randint(0, 100) for _ in pets]
#print("Pets:", pets)
#print("Data:", data)

import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))  
plt.pie(data, labels=pets, autopct='%1.2f%%', startangle=140)
plt.title('What is your favourite pet?')
```

### Bar Graphs 
<a id="bar-graphs"></a>

Bar graphs represent data using rectangular bars of different lengths. This is useful for comparing data from different categories.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))  
plt.bar(pets, data, color='skyblue')

plt.xlabel('Pet Types')  # Label for the x-axis
plt.ylabel('Values')  # Label for the y-axis
plt.title('Distribution of Pets')
```
